```python
df = pd.DataFrame([[1, 3, 5], [2, 4, 6], [2, 5, 7]], index = ['x','y','z'], columns = ['a','b','c'])
print(df)
print(df.at['x','b'])
print(df.iat[0, 1])
print(df.loc['x'])
print(df.loc[['x','y']])
print(df.loc['y':'z', 'b':'c'])
print(df.iloc[0])
print(df['a'])
print(df['a']['x'])
print(df[['a','c']])
print(df[:2])
print(df[df['b'] > 3])
```

```python
scores = pd.DataFrame(np.array([[80, 75, 88, 80, 78], [88, 86, 90, 95, 86],
[92, 85, 92, 98, 90], [81, 88, 80, 82, 85], [75, 80, 78, 80, 70]]))

print(scores)
print(scores.mean(axis = 1))
print(scores.median(axis = 1))
print(scores.mean(axis = 0))
print(scores.median(axis = 0))
scores.insert(0, column = "name", value = ["學生1", "學生2", "學生3", "學生4", "學生5"])
scores.insert(6, column = "mean", value = scores.iloc[:, 1:6].mean(axis = 1))
scores.insert(7, column = "median", value = scores.iloc[:, 1:6].median(axis = 1))
print(scores)
scores[['name','mean']].plot(kind = 'line', x = 'name', legend = False)
```
